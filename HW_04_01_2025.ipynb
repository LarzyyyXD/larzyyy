 Задание 1 <br>
 Создайте класс Circle (окружность). Для данного <br>
 класса реализуйте ряд перегруженных операторов: <br>
 ■ Проверка на равенство радиусов двух окружностей <br>
 (операция = =); <br>
 ■ Сравнения длин двух окружностей (операции >, <, <br>
 <=,>=); <br>
 ■ Пропорциональное изменение размеров окружности, <br>
 путем изменения ее радиуса (операции + - += -=). <br>


```python

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def __eq__(self, other):
        if isinstance(other, Circle):
            return self.radius == other.radius
        return NotImplemented

    def __lt__(self, other):
        if isinstance(other, Circle):
            return self.radius < other.radius
        return NotImplemented

    def __gt__(self, other):
        if isinstance(other, Circle):
            return self.radius > other.radius
        return NotImplemented

    def __le__(self, other):
        if isinstance(other, Circle):
            return self.radius <= other.radius
        return NotImplemented

    def __ge__(self, other):
        if isinstance(other, Circle):
            return self.radius >= other.radius
        return NotImplemented

    def __add__(self, other):
        if isinstance(other, (int, float)):
            return Circle(self.radius + other)
        return NotImplemented

    def __sub__(self, other):
        if isinstance(other, (int, float)):
            return Circle(self.radius - other)
        return NotImplemented

    def __iadd__(self, other):
        if isinstance(other, (int, float)):
            self.radius += other
            return self
        return NotImplemented

    def __isub__(self, other):
        if isinstance(other, (int, float)):
            self.radius -= other
            return self
        return NotImplemented



circle1 = Circle(5)
circle2 = Circle(3)

circle1 += 2  
circle2 -= 1  

print(f'Радиус circle1: {circle1.radius}')  
print(f'Радиус circle2: {circle2.radius}')  

print(circle1 > circle2)
```

Задание 3 <br>
Вам небходимо создать класс Airplane (самолет). <br>
 помощию перезагрузки операторов реализовать: <br>
■ Проверка на равенство типов самолетов (операция <br>
= =); <br>
■ Увеличение и уменьшение пассажиров в салоне са- <br>
молета (операции + - += -=); <br>
■ Сравнение двух самолетов по максимально возмож- <br>
ному количеству пассажиров на борту (операции > <br>
< <= >=). <br>

```python

class Airplane:
    def __init__(self, type_airplane, passengers, max_passengers):
        self.type_airplane = type_airplane
        self.passengers = passengers
        self.max_passengers = max_passengers
    
    def __eq__(self, other):
        if isinstance(other, Airplane):
            return self.type_airplane == other.type_airplane
        return NotImplemented

    def __lt__(self, other):
        if isinstance(other, Airplane):
            return self.max_passengers < other.max_passengers
        return NotImplemented

    def __gt__(self, other):
        if isinstance(other, Airplane):
            return self.max_passengers > other.max_passengers
        return NotImplemented

    def __le__(self, other):
        if isinstance(other, Airplane):
            return self.max_passengers <= other.max_passengers
        return NotImplemented

    def __ge__(self, other):
        if isinstance(other, Airplane):
            return self.max_passengers >= other.max_passengers
        return NotImplemented

    def __add__(self, other):
        if isinstance(other, (int, float)):
            return Airplane(self.passengers + other)
        return NotImplemented

    def __sub__(self, other):
        if isinstance(other, (int, float)):
            return Airplane(self.passengers - other)
        return NotImplemented

    def __iadd__(self, other):
        if isinstance(other, (int, float)):
            self.passengers += other
            return self
        return NotImplemented

    def __isub__(self, other):
        if isinstance(other, (int, float)):
            self.passengers -= other
            return self
        return NotImplemented

airplane1 = Airplane("Boeing 737", 150, 200)
airplane2 = Airplane("Airbus A320", 180, 220)

print(airplane1 == airplane2)  
print(airplane1 < airplane2)   
airplane1 += 10
print(airplane1.passengers)
```

Задание 4 <br>
Создать класс Flat (квартира). Реализовать перегру- <br>
женные операторы: <br>
■ Проверка на равенство площадей квартир (операция <br>
==); <br>
■ Проверка на неравенство площадей квартир (опера- <br>
ция !=); <br>
■ Сравнение двух квартир по цене (операции > < <= >=). <br>


```python

class Flat:
    def __init__(self, square, price):
        self.square = square
        self.price = price
    
    def __eq__(self, other):
        if isinstance(other, Flat):
            return self.square == other.square
        return NotImplemented

    def __ne__(self, other):
        return not self == other

    def __lt__(self, other):
        if isinstance(other, Flat):
            return self.price < other.price
        return NotImplemented

    def __le__(self, other):
        if isinstance(other, Flat):
            return self.price <= other.price
        return NotImplemented

    def __gt__(self, other):
        if isinstance(other, Flat):
            return self.price > other.price
        return NotImplemented

    def __ge__(self, other):
        if isinstance(other, Flat):
            return self.price >= other.price
        return NotImplemented

    def __add__(self, amount):
        if isinstance(amount, (int, float)):
            return Flat(self.square, self.price + amount)
        return NotImplemented

    def __sub__(self, amount):
        if isinstance(amount, (int, float)):
            new_price = self.price - amount
            if new_price < 0:
                return Flat(self.square, new_price)
        return NotImplemented

    def __iadd__(self, amount):
        if isinstance(amount, (int, float)):
            self.price += amount
            return self
        return NotImplemented

    def __isub__(self, amount):
        if isinstance(amount, (int, float)):
            new_price = self.price - amount
            if new_price < 0:
                self.price = new_price
            return self
        return NotImplemented

    def __repr__(self):
        return f"Flat(square={self.square}, price={self.price})"

flat1 = Flat(50, 100000)
flat2 = Flat(75, 150000)

print(flat1 == flat2)
print(flat1 != flat2)
print(flat1 < flat2)


flat1 += 5000
print(flat1)

flat1 -= 10000
print(flat1)
```
