Написать программу «справочник».Создать два списка <br>
целых. Один список хранит идентификационные коды, <br>
второй — телефонные номера. Реализовать меню для <br>
пользователя:<br> 
■ Отсортировать по идентификационным кодам; <br>
■ Отсортировать по номерам телефона; <br>
■ Вывести список пользователей с кодами и телефонами; <br>
■ Выход.

```python
def display_menu():
    print("\nМеню:")
    print("1. Отсортировать по идентификационным кодам")
    print("2. Отсортировать по номерам телефона")
    print("3. Вывести список пользователей с кодами и телефонами")
    print("4. Выход")

def sort_by_ids(ids, phones):
    combined = list(zip(ids, phones))
    combined.sort()  # Сортировка по первому элементу (идентификационные коды)
    return zip(*combined)  # Разделяем обратно на два списка

def sort_by_phones(ids, phones):
    combined = list(zip(ids, phones))
    combined.sort(key=lambda x: x[1])  # Сортировка по второму элементу (номера телефонов)
    return zip(*combined)  # Разделяем обратно на два списка

def main():
    ids = [101, 102, 103, 104]
    phones = ["123-456-7890", "987-654-3210", "555-555-5555", "111-222-3333"]

    while True:
        display_menu()
        choice = input("Выберите опцию (1-4): ")

        if choice == '1':
            ids, phones = sort_by_ids(ids, phones)
            print("Список отсортирован по идентификационным кодам.")
        elif choice == '2':
            ids, phones = sort_by_phones(ids, phones)
            print("Список отсортирован по номерам телефона.")
        elif choice == '3':
            print("\nСписок пользователей:")
            for id, phone in zip(ids, phones):
                print(f"ID: {id}, Телефон: {phone}")
        elif choice == '4':
            print("Выход из программы.")
            break
        else:
            print("Неверный выбор. Пожалуйста, попробуйте снова.")

if __name__ == "__main__":
    main()
```
