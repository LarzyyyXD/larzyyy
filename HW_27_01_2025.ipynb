Пользователь вводит с клавиатуры набор чисел. Полученные числа необходимо сохранить в список (тип <br>
списка нужно выбрать в зависимости от поставленной <br>
ниже задачи). После чего нужно показать меню, в котором <br>
предложить пользователю набор пунктов: <br>
1. Добавить новое число в список (если такое число существует в списке, нужно вывести сообщение пользователю об этом, без добавления числа).
2. Удалить все вхождения числа из списка (пользователь
вводит с клавиатуры число для удаления).
3. Показать содержимое списка (в зависимости от выбора пользователя список нужно показать с начала
или с конца).
4. Проверить есть ли значение в списке.
5. Заменить значение в списке (пользователь определяет заменить ли только первое вхождение или все
вхождения). <br>
В зависимости от выбора пользователя выполняется <br>
действие, после чего меню отображается снова. <br>

```python
user_list = input('Введите числа разделённые пробелом:')
lst = []
for i in user_list.split():
        lst.append(int(i))
while True:

    
    print(lst)
    print(
'1. Добавить новое число в список.\n\
2. Удалить все вхождения числа из списка.\n\
3. Показать содержимое списка.\n\
4. Проверить есть ли значение в списке.\n\
5. Заменить значение в списке.\n\
6. Выход.')
    user_menu = input('Ваш выбор: ')
    if user_menu == '1':
        user_input = int(input('Введите число для добавления: '))

        if user_input in lst:
            print('Такое число уже существует.')
        else:
            lst.append(user_input)

    elif user_menu == '2':
        user_input = int(input('Введите число для удаления: '))

        if user_input in lst:
            for i in lst:
                lst.remove(user_input)

        else:
            print('Такого числа не существует.')

    elif user_menu == '3':
        user_input = input('1. Показать список с начала.\n2. Показать список с конца. ')

        if user_input == '1':
            print('Список с начала:',sorted(lst))

        elif user_input == '2':
            print('Список с конца:',sorted(lst, reverse=True))
        
    elif user_menu == '4':
        user_input = int(input('Есть ли данное значение в списке: '))

        if user_input in lst:
            print('да')

        else:
            print('нет')

    elif user_menu == '5':
        user_num = int(input('Введите число которое хотите заменить: '))
        user_replace = int(input('Введите на что хотите заменить: '))
        user_input = input('1. Заменить только 1 вхождение.\n2.Заменить все вхождения. ')
        
        if user_input == '1':
            for i, value in enumerate(lst):
                if value == user_num:
                    lst[i] = user_replace
                    break

        elif user_input == '2':
            for i, value in enumerate(lst):
                if value == user_num:
                    lst[i] = user_replace

    elif user_menu == '6':
        break
                
    print(lst)
    
      
```

```python
class Stack:
    def __init__(self):
         self.items = []

    def isEmpty(self):
         return self.items == []

    def push(self, item):
         self.items.append(item)

    def pop(self):
         return self.items.pop()

    def peek(self):
         return self.items[len(self.items)-1]

    def size(self):
         return len(self.items)
    
    def clear_stack(self):
         return self.items.clear()

s = Stack()
print(s.isEmpty())
s.push('dog')
s.push(4)
print(s.peek())
print(s.size())
s.pop()
print(s.peek())
print(s.isEmpty())
s.clear_stack()
def display_menu():
    print("\nМеню:")
    print("1. Добавить строку в стек")
    print("2. Удалить строку из стека")
    print("3. Посчитать количество строк в стеке")
    print("4. Проверить пустой ли стек")
    print("5. Проверить полный ли стек")
    print("6. Очистить стек")
    print("7. Получить верхнюю строку без удаления")
    print("0. Выход")

# в процессе понимания как это делать
```
